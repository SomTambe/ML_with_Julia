```julia
using Plots, Zygote
```

```julia
x = collect(1:0.5:50)
y = map(x->x^2 + 2x + 1, x)
plot(x,y,seriestype = :scatter)
```

```julia
f(x,a,b,c)=a*x^2+b*x+c
loss(y,t) = (y-t)^2/2
```

```julia
a = rand(); b = rand(); c=rand();
 = 1e-6
while true
    net_loss = 0
    grad_a = 0
    grad_b = 0
    grad_c = 0
    for i = 1:length(x)
        t = f(x[i], a, b, c)
        net_loss += loss(t, y[i])
        grad_a += x[i]*x[i]*(y[i]-t)
        grad_b += x[i]*(y[i]-t)
        grad_c += y[i]-t
    end
    a = a + *grad_a/length(x)
    b = b + *grad_b/length(x)
    c = c + *grad_c/length(x)
    net_loss = net_loss/length(x)
    println("Loss: ",net_loss)
    if net_loss < 0.5
        break
    end
end
```

```julia
function forward(x,y,a,b,c)
    y = f(x,a,b,c)
    l = loss(y,y)
    return l
end
```

```julia
a = rand(); b = rand(); c=rand();
 = 1e-6
while true
    net_loss = 0
    grad_a = 0
    grad_c = 0
    grad_b = 0
    for i = 1:length(x)
        net_loss += forward(x[i],y[i],a,b,c)
        _,_,g_a,g_b,g_c = gradient((a,b,c,d,e)->forward(a,b,c,d,e),x[i],y[i],a,b,c)
        grad_a += g_a
        grad_b += g_b
        grad_c += g_c
    end
    a = a - *grad_a/length(x)
    b = b - *grad_b/length(x)
    c = c - *grad_c/length(x)
    net_loss = net_loss/length(x)
    println("Loss: ", net_loss)
    if net_loss < 1.5
        break
    end
end
```
