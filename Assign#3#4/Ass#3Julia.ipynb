```julia

using Plots, Zygote
```

```julia
x = collect(1:10:50)
y = map(x->4x^2+3x+4, x)
plot(x,y,seriestype = :scatter)
```

```julia
f(x,w2,w₁,b₁) = w2*x^2+w₁*x + b₁
loss(y,t) = (y-t)^2/2
```

```julia
# USING HAND WRITTEN DERIVATIVES
w2=rand();w1 = rand(); b =rand();

η = 10^-6
while true
    net_loss = 0
    grad_w2=0
    grad_w1 = 0
    grad_b = 0
    for i = 1:length(x)
        # loss calculation
        t = f(x[i], w2,w1, b)
        net_loss += loss(t, y[i])
        grad_w2+=x[i]^2*(y[i]-t)
        grad_w1 += x[i]*(y[i]-t)
        grad_b += y[i]-t

    end

       w2 = w2 + η*grad_w2/length(x)
    w1 = w1 + η*grad_w1/length(x)
    b = b +η*grad_b/length(x)
    net_loss = net_loss/length(x)
    println("Loss: ",net_loss)
    if net_loss < 40
        break
    end
end
```

```julia
w2,w1,b
```

```julia
z= collect(1:0.5:50) #array made more dense for smooth output when plotted
```

```julia
plot!(z,f.(z,w2,w1,b))
```

```julia
function forward(x,y,w2,w1,b)
    ŷ = f(x,w2,w1,b)
    l = loss(y,ŷ)
    return l
end
```

```julia
#USING ZYGOTE
w1= 2;w2=2; b =2;
η = 10^-6
while true
    net_loss = 0
    grad_w2 = 2
    grad_w1 = 2
    grad_b = 1
    for i = 1:length(x)
        net_loss += forward(x[i],y[i],w2,w1,b)
       _,_,g_w2,g_w1,g_b = gradient((a,b,c,d,e)->forward(a,b,c,d,e),x[i],y[i],w2,w1,b)
        grad_w1 += g_w1
        grad_w2 += g_w2
        grad_b += g_b
    end
    w2 = w2 - η*grad_w2/length(x)
    w1 = w1 - η*grad_w1/length(x)
    b = b - η*grad_b/length(x)
    net_loss = net_loss/length(x)
    println("Loss: ", net_loss)
    if net_loss < 5.5
        break
    end
end
```
