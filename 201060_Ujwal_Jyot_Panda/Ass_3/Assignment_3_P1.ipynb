```julia
using Plots
```

```julia
x = collect(1:0.5:50)
y = map(x->x^2+3x+4, x)
plot(x,y,seriestype = :scatter)
```

```julia
f(x,a,b,c) = a*(x^2)+b*x+c
loss(y,t) = (y-t)^2/2
```

```julia
a = rand(); b = rand(); c = rand();
 = 1e-6
while true
    net_loss = 0
    grad_a = 0
    grad_b = 0
    grad_c = 0
    for i = 1:length(x)
        
        t = f(x[i], a, b, c)
        net_loss += loss(t, y[i])
        grad_a += (x[i]^2)*(y[i]-t)
        grad_b += (y[i]-t)*x[i]
        grad_c += (y[i]-t)
        
    end
    
    a = a + *grad_a/length(x)
    b = b + *grad_b/length(x)
    c = c + *grad_c/length(x)
    net_loss = net_loss/length(x)
    println("Loss: ",net_loss)
    if net_loss < 0.05
        break
    end
end
```
