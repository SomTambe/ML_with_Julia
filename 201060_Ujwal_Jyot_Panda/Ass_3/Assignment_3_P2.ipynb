```julia
using Plots, Zygote
```

```julia
x = collect(1:0.5:50)
y = map(x->x^2+3x+4, x)
plot(x,y,seriestype = :scatter)
```

```julia
f(x,a,b,c) = a*(x^2)+b*x+c
loss(y,t) = (y-t)^2/2
```

```julia
function forward(x,y,a,b,c)
    ŷ = f(x,a,b,c)
    l = loss(y,ŷ)
    return l
end
```

```julia
a = rand(); b = rand(); c = rand();
η = 1e-6
while true
    net_loss = 0
    grad_a = 0
    grad_b = 0
    grad_c = 0
    for i = 1:length(x)
        net_loss += forward(x[i],y[i],a,b,c)
        _,_,g_a,g_b,g_c = gradient((a,b,c,d,e)->forward(a,b,c,d,e),x[i],y[i],a,b,c)
        grad_a += g_a/2
        grad_b += g_b
        grad_c += g_c
    end
    a = a - η*grad_a/length(x)
    b = b - η*grad_b/length(x)
    c = c - η*grad_c/length(x)
    net_loss = net_loss/length(x)
    println("Loss: ", net_loss)
    if net_loss < 1.5
        break
    end
end
```
