```julia
using Plots, Zygote
```

Let the quadratic expression be $f(x) = 2x^2 + 3x + 1$

```julia
x = collect(1:0.1:10)
y = map(x-> 2x^2 + 3x + 1, x)
```

```julia
plot(x,y, seriestype=:scatter)
```

```julia
f(x,a1,b1,c1) = a1*(x^2) + b1*x + c1   #
```

```julia
println(f(1.0,1.0,1.0,1.0))
```

```julia
loss(y,t) = ((y-t)^2)/2
```

```julia
a = 0

b = 0

c = 0
```

## Batch Gradient Descent

```julia
lr=1e-4
epochs = 500
i=0
while true
    total_loss = 0
    da = 0
    db = 0
    dc = 0
    for j in 1:length(x)
        total_loss+=loss(f(x[j],a,b,c),y[j])
        da+=((x[j]^2)*(f(x[j],a,b,c)-y[j]))
        db+=(x[j]*(f(x[j],a,b,c)-y[j]))
        dc+=(f(x[j],a,b,c)-y[j])
    end
    total_loss/=length(x)
    da/=length(x)
    db/=length(x)
    dc/=length(x)
    a = a-lr*da
    b = b-lr*db
    c = c-lr*dc
    if i%5==0
        print(i)
        println("  Loss = ", total_loss)
    end
    i+=1
    if total_loss < 0.0005
        break
    end
end
```

```julia
println("a = ", a)
println("b = ", b)
println("c = ", c)
```

```julia
a=0
b=0
c=0
```

## Stochastic Gradient Descent

```julia
lr=1e-5
epochs = 500
i=0
while true
    total_loss = 0
    da = 0
    db = 0
    dc = 0
    for j in 1:length(x)
        total_loss=loss(f(x[j],a,b,c),y[j])
        da=((x[j]^2)*(f(x[j],a,b,c)-y[j]))
        db=(x[j]*(f(x[j],a,b,c)-y[j]))
        dc=(f(x[j],a,b,c)-y[j])
        a = a-lr*da
        b = b-lr*db
        c = c-lr*dc
    end
    if i%5==0
        println("Loss = ", total_loss)
    end
    i+=1
    if total_loss < 0.00001
        break
    end
end
```

```julia
println("a = ", a)
println("b = ", b)
println("c = ", c)
```

```julia
plot!(x, f.(x,a,b,c))
```

## Using Zygote (automatic differentiation)

```julia
a=0
b=0
c=0
```

```julia
lr = 1e-4
i=0
while true
    l = 0
    da=0
    db=0
    dc=0
    for j in 1:length(x)
        l = loss(f(x[j],a,b,c),y[j])
        _,da,db,dc,_ = gradient((p,q,r,s,t)->loss(f(p,q,r,s),t),x[j],a,b,c,y[j])
        a = a-lr*da
        b = b-lr*db
        c = c-lr*dc
        
        print(i)
        println("  Loss - ", l)
        
    end
    if i%5==0
        print(i)
        println("  Loss - ", l)
    end
    i+=1
    if l<0.00001
        break
    end
end
```

```julia
println("a = ", a)
println("b = ", b)
println("c = ", c)
```
