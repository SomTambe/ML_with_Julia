```julia
using Plots, Zygote, Flux
```

```julia
f(a, b, c, x)=a*x^2+b*x+c
loss(y,t)=(y-t)^2/2
```

```julia
function forward(x,y,a,b,c)
    ŷ = f(a, b, c, x)
    l = loss(y,ŷ)
    return l
end
```

```julia
x = collect(0:0.5:50)
y = map(x->x^2+5*x+10, x)
plot(x,y,seriestype = :scatter)
```

```julia
a=rand();
b=rand();
c=rand();
η=1e-6
while true
    net_loss=0
    grad_a=0
    grad_b=0
    grad_c=0
    for i=1:length(x)
        net_loss += forward(x[i],y[i],a,b,c)
        _,_,g_a,g_b,g_c = gradient((a,b,c,d,e)->forward(a,b,c,d,e),x[i],y[i],a,b,c)
        grad_a += g_a
        grad_b += g_b
        grad_c += g_c
    end
    a = a - η*grad_a/length(x)
    b = b - η*grad_b/length(x)
    c = c - η*grad_c/length(x)
    net_loss = net_loss/length(x)
    println("Loss: ", net_loss)
    if net_loss < 90
        break
    end
end
```
