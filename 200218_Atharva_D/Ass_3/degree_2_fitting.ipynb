# Use Zygote and Hand-Written derivative rules to fit over a 2-Degree Polynomial Curve

```julia
using Plots, Zygote
```

### Choosing the quadratic function and plotting it

```julia
x = collect(1:0.5:50)
y = map(x->x^2+2x+1, x)
plot(x,y,seriestype = :scatter)
```

### Writing a general quadratic function with paramters a1, b1 and c1 and defining Loss function

```julia
f(x, a1, b1, c1) = a1*x^2 + b1*x + c1
loss(y,t) = (y-t)^2/2
```

```julia
a = 0.5; b=1.5; c= 1.5;
a, b, c
```

```julia
a = rand(); b = rand(); c = rand();
 = 1e-6
while true
    net_loss = 0
    grad_a = 0
    grad_b = 0
    grad_c = 0
    
    for i = 1:length(x)
        # loss calculation
        t = f(x[i], a, b, c)
        net_loss += loss(t, y[i])
        grad_a += (x[i]^2)*(y[i]-t)
        grad_b += x[i]*(y[i]-t)
        grad_c += (y[i]-t)

    end
    # updation
    a = a + *grad_a/length(x)
    b = b + *grad_b/length(x)
    c = c + *grad_c/length(x)
    net_loss = net_loss/length(x)
    println("Loss: ",net_loss)
    if net_loss < 0.0005
        break
    end
end
```
