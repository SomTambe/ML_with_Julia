# Polynomial Regression with degree 2

```julia
using Plots,Zygote
```

Approximating the function to f(x)=$4x^2 + 6x + 9$

```julia
begin
		x = collect(1:0.5:50)
		y = map(x->4x^2 + 6x + 9,x) #for DataSet
		plot(x,y,seriestype = :scatter)
end
```

```julia
f(x,a,b,c) = a*x^2 + b*x + c
```

```julia
loss(a,b) = (a-b)^2/2
```

```julia
begin
	a=rand();b=rand();c=rand()
	=1e-6
	while true
		net_loss = 0
		grad_a = 0
		grad_b = 0
		grad_c = 0
		for i = 1:length(x)
			t = f(x[i],a,b,c)
			net_loss += loss(t,y[i])
			grad_a = x[i]*x[i]*(y[i]-t)
            grad_b = x[i]*(y[i]-t)
            grad_c = y[i]-t
		end
		a = a + *grad_a/length(x)
		b = b + *grad_b/length(x)
		c = c + *grad_c/length(x)
		net_loss /= length(x)
		println("Loss: ",net_loss)
		if net_loss < 0.5
			break
		end
	end
end
```

```julia
plot!(x,f.(x,a,b,c))
```

```julia
function forward(x,y,a,b,c)
	y = f(x,a,b,c)
	l = loss(y,y)
	return l
end
```

```julia
gradient((l,m,n,p)->f(l,m,n,p),1,2,3,4)
```

```julia
begin
	a=rand();b=rand();c=rand()
	=1e-6
	while true
		net_loss = 0
		grad_a = 0
		grad_b = 0
		grad_c = 0
		for i = 1:length(x)
			t = f(x[i],a,b,c)
			net_loss += loss(t,y[i])
			_,_,g_a,g_b,g_c = gradient((l,m,n,p,q)->forward(l,m,n,p,q),x[i],y[i],a,b,c)
			grad_a += g_a
			grad_b += g_b
			grad_c += g_c
		end
		a = a - *grad_a/length(x)
		b = b - *grad_b/length(x)
		c = c - *grad_c/length(x)
		net_loss /= length(x)
		println("Loss: ",net_loss)
		if net_loss < 0.05
			break
		end
	end
end
```
